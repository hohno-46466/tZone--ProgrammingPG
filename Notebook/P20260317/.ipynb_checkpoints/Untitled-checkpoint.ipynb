```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
c = np.arange(-20, 101, 5)
```

```python
print(c)
```

```python
f = 9/5 * c + 32
```

```python
print(f)
```

```python
df = pd.DataFrame({"Celsius": c, "Fahrenheit": f})
```

```python
print(df)
```

```python
df2 = pd.read_csv("data.csv")
print(df2)
```

```python
plt.figure(figsize=(6,4))
plt.plot(df2["Celsius"], df2["Fahrenheit"])
plt.xlabel("Celsius (°C)")
plt.ylabel("Fahrenheit (°F)")
plt.title("Celsius vs Fahrenheit")
plt.grid()
plt.show()
```
