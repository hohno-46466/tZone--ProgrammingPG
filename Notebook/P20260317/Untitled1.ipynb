```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
salary = np.linspace(0, 1000000, 500)
```

```python
pension = 200000
threshold = 650000
```

```python
reduction = np.maximum(0, (salary + pension - threshold)/2)
```

```python
pension_after = pension - reduction
```

```python
total_income = salary + pension_after
```

```python
print(total_income)
```

```python
plt.figure(figsize=(6,4))

plt.plot(salary, total_income)

plt.xlabel("Salary (yen/month)")
plt.ylabel("Total income after pension reduction")

plt.title("Working while receiving pension")

plt.grid()

plt.show()
```
