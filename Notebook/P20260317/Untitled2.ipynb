```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams["font.family"] = "Hiragino Sans"
plt.rcParams["axes.unicode_minus"] = False

# 給与の範囲（円/月）
salary = np.linspace(0, 1200000, 500)

# 条件設定（円/月）
pension = 200000
threshold = 650000

# 減額額
reduction = np.maximum(0, (salary + pension - threshold) / 2)

# 減額後年金（0未満にはしない）
pension_after = np.maximum(0, pension - reduction)

# 合計収入
total_income = salary + pension_after

# 参考線
total_no_reduction = salary + pension   # 減額なし
total_salary_only = salary              # 年金なし

# 万円単位に変換
salary_man = salary / 10000
income_man = total_income / 10000
income_no_reduction_man = total_no_reduction / 10000
income_salary_only_man = total_salary_only / 10000

# 重要な点
start_reduction = threshold - pension   # 減額開始給与
end_pension = threshold + pension       # 年金ゼロ到達給与

start_reduction_man = start_reduction / 10000
end_pension_man = end_pension / 10000

# グラフ
plt.figure(figsize=(8, 6))

plt.plot(salary_man, income_man, linewidth=2, label="減額後の合計収入")
plt.plot(salary_man, income_no_reduction_man, "--", linewidth=2, label="減額なしの場合")
plt.plot(salary_man, income_salary_only_man, ":", linewidth=2, label="年金なし（給与＝収入）")

# 縦線
plt.axvline(start_reduction_man, color="gray", linestyle=":", linewidth=1.5)
plt.axvline(end_pension_man, color="gray", linestyle=":", linewidth=1.5)

plt.scatter([start_reduction_man], [(start_reduction + pension) / 10000], color="black")
plt.scatter([end_pension_man], [end_pension_man], color="black")

# 注記
plt.text(start_reduction_man + 1, pension / 10000 + start_reduction_man + 2,
         "減額開始", fontsize=10)
plt.text(end_pension_man + 1, end_pension_man + 2,
         "年金 0", fontsize=10)

plt.xlabel("給与（万円/月）")
plt.ylabel("合計収入（万円/月）")
plt.title("在職老齢年金の簡易モデル")
plt.xlim(0, 120)
plt.ylim(0, 130)
plt.grid()
plt.legend()
plt.show()
```
